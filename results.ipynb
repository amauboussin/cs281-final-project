```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
None
```

```python
def plot(data):
    order = ['Random Forest', 'LDA (K=40)', 'hDP']
    keys, values = zip(*sorted(data.items(), key=lambda x: order.index(x[0]), reverse=False))
    
    width = .75
    x = np.arange(len(data))
    plt.bar(x, values, color='steelblue')
    plt.xticks(x + width / 2, keys)
```

```python
data = {
'Random Forest' : .555,
'hDP' : 996/1499.,
'LDA (K=40)': 577 / 1499.,
}
plt.figure(figsize=(10,8))

plot(data)
plt.title('Thirty Subreddit Classification ')
plt.xlabel('Method')
plt.ylabel('Classification Accuracy')
plt.show()
```
