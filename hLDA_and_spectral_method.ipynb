```python
import gensim as gs
import numpy as np
import matplotlib.pyplot as plt
from nltk import word_tokenize, download
from nltk.corpus import stopwords
from load import comments_list, comments_grouped_by
import logging
logger = logging.getLogger()
logger.setLevel('INFO')
```

```python
comments = comments_list()
```

```python
bodies = []
for comment in comments:
    bodies.append(comment["body"])
```

```python
token_lists = []
for body in bodies:
    tokens = word_tokenize(body.lower())
    tokens = [w for w in tokens if w not in stopwords.words('english') and w.isalnum()]
    token_lists.append(tokens)
```

```python
dictionary = gs.corpora.Dictionary(token_lists)
dictionary.save('comment_tokens.dict')
corpus = [dictionary.doc2bow(token_list) for token_list in token_lists]
```

```python
hdp = gs.models.HdpModel(corpus, dictionary, T=40)
```

```python
hdp.evaluate_test_corpus([[(1,1), (2,1)]])
```
