```python
from collections import Counter

from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.util import ngrams

from load import comments_list, comments_grouped_by
```

```python
to_remove = set(['http', 'faq', 'https', 'amp','source', 'deletion', 'sfw',
              'nsfw', 'gt', 'gon', 'na', 'delete', 'comment', 'profile'])
def _filter(w):
    return all([w.isalnum(), w not in stopwords.words('english'), w not in to_remove])
```

```python
def most_common(comments, n, number=100):
    text = ' '.join([c['body'] for c in comments])
    tokens = word_tokenize(text)
    tokens = [w.lower() for w in tokens ]
    tokens = filter(_filter, tokens)
    bigrams= list(ngrams(tokens, n))
    return Counter(bigrams).most_common(number)
```
